```python
import pandas as pd

# Read the text file into a DataFrame
path_to_txt = "database/data/UIU_question_answer_raw.txt"
data = pd.read_csv(path_to_txt, sep="\t", names=['Q&A'])

data
```

```python
# Create two new columns for odd and even indexes
data['question'] = data['Q&A'][::2].reset_index(drop=True)  # odd indexes will be question
data['answer'] = data['Q&A'][1::2].reset_index(drop=True)   # even indexes will be answer

data
```

```python
data = data.drop('Q&A', axis=1)                 # drop the Q&A column
data
```

```python
print(data.isnull().sum())
data = data.dropna()
data
```

```python
data.to_csv('transcript_200.csv', index=False)  # Save the new dataframe into csv file 
data
```
